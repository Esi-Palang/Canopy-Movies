Data Cleaning and Modeling:

```python
#Create a new DataFrame Consisting these columns: ID,Title,Age,Genres,Langauge,IMDb
movie_nds= movie_ds1.loc[:, ['Title','Age','Genres','Language', 'IMDb', 'Rotten Tomatoes']]

#View the new DataFrame 
movie_nds.head()

```

```python
#Check for the missing data
movie_nds.isnull().sum()
```

```python
#Combine both datasets : movie_nds and movie_ds2 using merge() function and Title as Primary Key
movie_fds = pd.merge(movie_nds , movie_ds2,
                    on = 'Title')
#print the new dataframe
movie_fds.head()
```

```python
movie_fds.info()
```

```python
#Find and Deleting duplicates
clean_m_fds= movie_fds.drop_duplicates(subset = ('ID'))
#Delete Null values from IMDb
clean_m_fds = movie_fds.dropna(subset = ["IMDb"] )
#Print the shape
clean_m_fds.shape
```

```python
#Combine both datasets : movie_nds and movie_ds2 using merge() function and Title as Primary Key
movie_fds = pd.merge(movie_nds , movie_ds2,
                    on = 'Title')
#print the new dataframe
movie_fds.head()
```

```python
clean_m_fds.shape
```

Apparently there are no duplicates.

```python
#Find French language movies and drop the rest of rows that doesnt contain 'French' as a 'Language'
fmovie = movie_fds[movie_fds['Language'].str.contains(pat= 'French' , na = False)]
#print the result
fmovie.head()
```

Now its time to check which Genre has the lowest ranking.
Horror and Thriller have the most repeat in the lowest 20 range of rating.

```python
#Split the Genres values in French movies
genres=fmovie['Genres'].dropna().str.split(',')

#Count and print the values for Genres 
print(genres.value_counts())
```

```python
#Convert genres column values to row for counting the correct repetition of the values
genres= genres.explode()
#Count and print the values for Genres 
print(genres.value_counts())
```

```python
#Determine the average of each genres based on rating
#Call Genres and IMDb columns and drop the rest of the columns
genres_rating = fmovie[['Genres','IMDb']].dropna()

#Use Str.split() function to split the values in the column
genres_rating ['Genres']= genres_rating['Genres'].str.split(',')

#convert genres column values to row
genres_rating = genres_rating.explode('Genres')

#Use Groupby() function to Genres column values  distinctively
avg_g_r = genres_rating.groupby('Genres')

#calculate mean for IMDb rating and sort them Descending
avg_g_r = avg_g_r['IMDb'].mean().sort_values(ascending = False)
#print the outcome
print(avg_g_r)
```

According to the average IMDb rating for genres News and Animation and Documentary have the highest rate of rating.
Horror , Thriller and Action have the lowest rating.
